```python
arr = [i for i in range(10)]
arr = list(map(lambda x: x * 2, arr))
print(arr)
```

```python
# list的一些方法测试
import random
from functools import cmp_to_key
arr = [random.randint(0, 100) for _ in range(10)]
cmp = lambda x, y: x - y
arr.sort(key=cmp_to_key(cmp))
print(arr)
```

```python
arr = [
    (1, 'a'),
    (3, 'c'),
    (2, 'b')
]
arr.sort(key=lambda x: (x[0],x[1]))
print(arr)
```

```python
# python非零即True
arr = [-1, 0, 1, 2, 3, 4, 5]
for i in arr:
    if i:
        print(i)
```
